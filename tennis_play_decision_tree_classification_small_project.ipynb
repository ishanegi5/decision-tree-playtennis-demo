```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('PlayTennis.csv')
```

```python
dataset
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
dataset['Play Tennis']=le.fit_transform(dataset['Play Tennis'])
```

```python
ohe=OneHotEncoder(sparse_output=False)
encoded=ohe.fit_transform(dataset[['Outlook','Temperature','Humidity','Wind']])
df=pd.DataFrame(encoded,columns=ohe.get_feature_names_out(['Outlook','Temperature','Humidity','Wind']))
data=pd.concat([df,dataset[['Play Tennis']]],axis=1)
data
```

```python
x=data.drop('Play Tennis',axis=1)
```

```python
y=data['Play Tennis']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
dtc=DecisionTreeClassifier(random_state=42)
```

```python
params={'criterion':['entropy','gini'],'max_depth':[2,3,4],'min_samples_split':[2,3,4,5],'min_samples_leaf':[2,3,4,5]}
```

```python
grid=GridSearchCV(dtc,params,cv=4,scoring='accuracy')
```

```python
grid.fit(x_train,y_train)
```

```python
y_pred=grid.predict(x_test)
```

```python
print("Best parameters: ",grid.best_params_)
```

```python
print("Best score: ",grid.best_score_)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
plot_tree(grid.best_estimator_, filled=True, feature_names=['Outlook_Overcast','Outlook_Rain','Outlook_Sunny','Temperature_Cool','Temperature_Hot','Temperature_Mild','Humidity_High','Humidity_Normal','Wind_Strong','Wind_Weak'],class_names=['Yes','No'])
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(grid.best_estimator_,x,y,cv=4,scoring='accuracy')
```

```python
print(score)
```
